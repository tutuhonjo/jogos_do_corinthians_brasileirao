```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
navegador.get ('https://ge.globo.com/futebol/brasileirao-serie-a/')

navegador.maximize_window()
```

```python
from selenium.webdriver.common.by import By
import time

for i in range(39):
    tabela_jogos = navegador.find_elements(By.CLASS_NAME, 'lista-jogos__jogo')
    for jogo in tabela_jogos:
        if "COR" in jogo.text: 
            info = jogo.find_element(By.CLASS_NAME, 'jogo__informacoes')
            placar = jogo.find_element(By.CLASS_NAME, 'placar')
            print(f'DATA/LOCAL/HORA - {info.text} \n\nJOGO\n{placar.text}\n\n')
            txt = open('corinthians.txt', 'a')
            txt.write(f'{i+1}º Rodada\nDATA/LOCAL/HORA - {info.text} \n\nJOGO\n{placar.text}\n\n')
    navegador.find_element(By.XPATH, '//*[@id="classificacao__wrapper"]/section/nav/span[3]').click()
    time.sleep(0.5)
```

```python
navegador.refresh()
```
