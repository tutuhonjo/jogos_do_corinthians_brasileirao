```python
import pandas as pd
from selenium import webdriver
from datetime import date
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
navegador.get ('https://ge.globo.com/futebol/brasileirao-serie-a/')
```

```python
from selenium.webdriver.common.by import By
import time

jogos = []
datas = []
horas = []
locais = []

#For para quantidade de rodadas no Brasileirão
for i in range(38):
    #Localizar jogos da rodada
    tabela_jogos = navegador.find_elements(By.CLASS_NAME, 'lista-jogos__jogo')
    for jogo in tabela_jogos:
        #Condição para buscar jogos do corinthians na rodada
        if "COR" in jogo.text: 
            #Armazenando número da rodada na variável "rodada"
            rodada = navegador.find_element(By.CLASS_NAME, 'lista-jogos__navegacao--rodada')

            #Armazenando informações do jogo na variábel "info"
            info = jogo.find_element(By.CLASS_NAME, 'jogo__informacoes')

            #Isolando em uma variável o local do jogo, armazenando na variável "info"
            local = info.find_element(By.CLASS_NAME, 'jogo__informacoes--local')

            #Isolando em uma variável a data do jogo, armazenando na variável "info"
            data = info.find_element(By.CLASS_NAME, 'jogo__informacoes--data')

            #Isolando em uma variável o hora do jogo, armazenando na variável "info"
            hora = info.find_element(By.CLASS_NAME, 'jogo__informacoes--hora')

            #Armazenando o placar na variável "placar"
            placar = jogo.find_element(By.CLASS_NAME, 'placar')

            #Printando os jogos encontrados
            print(f'DATA - {data.text} \nLocal - {local.text} \nHora - {hora.text}\n\nJOGO\n{placar.text}\n\n')

            #Armazenando detalhes no bloco de notas
            txt = open('corinthians.txt', 'a')
            txt.write(f'DATA - {data.text} \nLocal - {local.text} \nHora - {hora.text}\n\nJOGO\n{placar.text}\n\n')

            #Armazenando valores em uma lista
            jogos.append(placar.text)
            datas.append(data.text)
            horas.append(hora.text)
            locais.append(local.text)

    #Clicando em próxima rodada para seguir o WebScraping        
    navegador.find_element(By.XPATH, '//*[@id="classificacao__wrapper"]/section/nav/span[3]').click()
    time.sleep(0.5)

#Transformando listas do WebScraping em Dataframe
df = pd.DataFrame({'Data': datas, 'Hora': horas, 'Local':locais, 'Jogo':jogos})

#Código para inserir data atual no Excel
mask = '%d%m%Y'
hoje = date.today().strftime(mask)

#Extraindo Dataframe em Excel
df.to_excel(f'jogos do corinthians {hoje}.xlsx', index=False)
```
